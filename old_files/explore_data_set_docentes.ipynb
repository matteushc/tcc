## Filtrando os docentes, de acordo com os filtros informados pela documentação: 
- O filtro TP_TIPO_TURMA NOT IN (4,5) diz respeito às turmas de Atividade Complementar e de Atendimento Educacional Especializado (AEE), que não são contabilizadas.
- O filtro TP_TIPO_DOCENTE IN (1,5) diz respeito às funções de Docente e de Docente Titular coordenador de tutoria (de módulo ou disciplina) – EaD, exercidas na escola.

```python
docentes_co = dados_docentes_co.loc[(~dados_docentes_co['TP_TIPO_TURMA'].isin([4,5])) 
                      & (dados_docentes_co['TP_TIPO_DOCENTE'].isin([1,5]))]

print('Total dados_docentes_co: ' + str(dados_docentes_co['ID_DOCENTE'].count()))
print('Total docentes_co: ' + str(docentes_co['ID_DOCENTE'].count()))
```

- ### Analisando Escolaridade dos Docentes X Etapas de Ensino

```python
#Criação de funções para tratamento dos DataSets
def etapaEnsino(x):
    result = ''
    if x in [1,2,3]: 
        result = '01-Educação Infantil' 
    elif x in [4,5,6,7,14,15,16,17,18,56]: 
        result = '02-Educação Fundamental (Anos Iniciais)' 
    elif x in [8,9,10,11,12,13,19,20,21,22,23,24,41]: 
        result = '03-Educação Fundamental (Anos Finais)' 
    elif x in [25,26,27,28,29]: 
        result = '04-Ensino Médio (Propedêutico)' 
    elif x in [30,31,32,33,34]: 
        result = '05-Curso Técnico Integrado' 
    elif x in [35,36,37,38]: 
        result = '06-Ensino Médio Normal' 
    elif x in [65,69,70,72]: 
        result = '07-EJA (Ensino Fundamental)' 
    elif x in [71]: 
        result = '08-EJA (Ensino Médio)' 
    elif x in [65,73]: 
        result = '09-EJA Profissional (Ensino Fundamental)' 
    elif x in [67,74]: 
        result = '10-EJA Profissional (Ensino Médio)' 
    elif x in [39]: 
        result = '11-Curso Técnico (Concomitante)' 
    elif x in [40]: 
        result = '12-Curso Técnico (Subsequente)' 
    elif x in [64]: 
        result = '13-Curso Técnico Misto (Concomitante e Subsequente)' 
    elif x in [68]: 
        result = '14-Curso FIC Concomitante' 
    else: 
        result = x
                
    return result
```

```python
# Escolaridade dos Docentes - Dividindo a escolaridade em campos
docentes_co.loc[(docentes_co['TP_ESCOLARIDADE'] >= 1), 'IN_FUNDAMENTAL_COMPLETO'] = 1
docentes_co.loc[(docentes_co['TP_ESCOLARIDADE'] < 1), 'IN_FUNDAMENTAL_COMPLETO'] = 0
docentes_co['IN_FUNDAMENTAL_COMPLETO'] = docentes_co['IN_FUNDAMENTAL_COMPLETO'].astype(int)

docentes_co.loc[(docentes_co['TP_ESCOLARIDADE'] >= 3), 'IN_ENSINO_MEDIO_COMPLETO'] = 1
docentes_co.loc[(docentes_co['TP_ESCOLARIDADE'] < 3), 'IN_ENSINO_MEDIO_COMPLETO'] = 0
docentes_co['IN_ENSINO_MEDIO_COMPLETO'] = docentes_co['IN_ENSINO_MEDIO_COMPLETO'].astype(int)

docentes_co.loc[(docentes_co['TP_ESCOLARIDADE'] >= 3) & (docentes_co['TP_NORMAL_MAGISTERIO'] == 0), 
                'IN_ENSINO_MEDIO_PROPEDEUTICO_COMPLETO'] = 1
docentes_co.loc[(docentes_co['TP_ESCOLARIDADE'] < 3) | (docentes_co['TP_NORMAL_MAGISTERIO'] != 0), 
                'IN_ENSINO_MEDIO_PROPEDEUTICO_COMPLETO'] = 0
docentes_co['IN_ENSINO_MEDIO_PROPEDEUTICO_COMPLETO'] = docentes_co['IN_ENSINO_MEDIO_PROPEDEUTICO_COMPLETO'].astype(int)

docentes_co.loc[(docentes_co['TP_ESCOLARIDADE'] >= 3) & (docentes_co['TP_NORMAL_MAGISTERIO'].isin([1,3])), 
                'IN_ENSINO_MEDIO_NORMAL_COMPLETO'] = 1
docentes_co.loc[(docentes_co['TP_ESCOLARIDADE'] < 3) | (~docentes_co['TP_NORMAL_MAGISTERIO'].isin([1,3])), 
                'IN_ENSINO_MEDIO_NORMAL_COMPLETO'] = 0
docentes_co['IN_ENSINO_MEDIO_NORMAL_COMPLETO'] = docentes_co['IN_ENSINO_MEDIO_NORMAL_COMPLETO'].astype(int)

docentes_co.loc[(docentes_co['TP_ESCOLARIDADE'] >= 3) & (docentes_co['TP_NORMAL_MAGISTERIO'] == 2), 
                'IN_ENSINO_MEDIO_NORMAL_INDIGENA_COMPLETO'] = 1
docentes_co.loc[(docentes_co['TP_ESCOLARIDADE'] < 3) | (docentes_co['TP_NORMAL_MAGISTERIO'] != 2), 
                'IN_ENSINO_MEDIO_NORMAL_INDIGENA_COMPLETO'] = 0
docentes_co['IN_ENSINO_MEDIO_NORMAL_INDIGENA_COMPLETO'] = docentes_co['IN_ENSINO_MEDIO_NORMAL_INDIGENA_COMPLETO'].astype(int)

docentes_co.loc[(docentes_co['TP_ESCOLARIDADE'] >= 4), 
                'IN_NIVEL_SUPERIOR_COMPLETO'] = 1
docentes_co.loc[(docentes_co['TP_ESCOLARIDADE'] < 4), 
                'IN_NIVEL_SUPERIOR_COMPLETO'] = 0
docentes_co['IN_NIVEL_SUPERIOR_COMPLETO'] = docentes_co['IN_NIVEL_SUPERIOR_COMPLETO'].astype(int)

docentes_co.loc[(docentes_co['TP_ESCOLARIDADE'] >= 1) & 
                ((docentes_co['TP_SITUACAO_CURSO_1'] == 2) |
                 (docentes_co['TP_SITUACAO_CURSO_2'] == 2) |
                 (docentes_co['TP_SITUACAO_CURSO_3'] == 2)), 
                'IN_NIVEL_SUPERIOR_CURSANDO'] = 1
docentes_co.loc[(docentes_co['TP_ESCOLARIDADE'] < 1) |
                 ((docentes_co['TP_SITUACAO_CURSO_1'] != 2) &
                 (docentes_co['TP_SITUACAO_CURSO_2'] != 2) &
                 (docentes_co['TP_SITUACAO_CURSO_3'] != 2)), 
                'IN_NIVEL_SUPERIOR_CURSANDO'] = 0
docentes_co['IN_NIVEL_SUPERIOR_CURSANDO'] = docentes_co['IN_NIVEL_SUPERIOR_CURSANDO'].astype(int)

docentes_co['TP_ETAPA_ENSINO_DESCRITIVO'] = docentes_co['TP_ETAPA_ENSINO'].apply(etapaEnsino)


docentes_co_pivot = docentes_co.pivot_table(columns='TP_ETAPA_ENSINO_DESCRITIVO', 
                                            values=['IN_FUNDAMENTAL_COMPLETO', 
                                                    'IN_ENSINO_MEDIO_COMPLETO', 
                                                    'IN_ENSINO_MEDIO_PROPEDEUTICO_COMPLETO', 
                                                    'IN_ENSINO_MEDIO_NORMAL_COMPLETO',
                                                    'IN_ENSINO_MEDIO_NORMAL_INDIGENA_COMPLETO', 
                                                    'IN_NIVEL_SUPERIOR_CURSANDO',
                                                    'IN_NIVEL_SUPERIOR_COMPLETO'], aggfunc=lambda x: round((x.sum() * 100) / x.count(), 2) )

docentes_co_pivot
```
