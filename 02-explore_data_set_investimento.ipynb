> ## Dados Investimento Municipio:<br>
- Base de Investimentos dos Municipios 
- SIOPE (Sistema de Informações sobre Orçamentos Públicos em Educação)
- https://www.fnde.gov.br/index.php/fnde_sistemas/siope/relatorios/arquivos-dados-analiticos
- Download: ftp://ftp2.fnde.gov.br/dadosabertos/SIOPE/DADOS_CONSOLIDADOS_2017.CSV

```python
investimento_edu = pd.read_csv(f'{diretorio_datasets}/DADOS_CONSOLIDADOS_2017.CSV', 
                            delimiter=";", encoding = "ISO-8859-1", decimal=",")
```

```python
investimento_edu.head()
```

```python
investimento_edu.info()
```

```python
columns_selected = ['CO_MUNICIPIO_IBGE', 'TP_PERIODO', 'VL_RECEITA_REALIZADA', 
                             'VL_RECEITA_ORCADA', 'VL_DESPESA_EMPENHADA_EDUCACAO', 
                             'VL_DESPESA_LIQUIDADA_EDUCACAO','VL_DESPESA_PAGA_EDUCACAO', 
                             'VL_DESPESA_ORCADA_EDUCACAO']

investimento_edu = investimento_edu[columns_selected]
investimento_edu.head()
```

```python
investimento_edu.describe()
```

```python
investimento_edu.isnull().any()
```

```python
investimento_edu.dropna(inplace=True)
investimento_edu.reset_index(drop=True, inplace=True)
```

```python
investimento_edu.info()
```

```python
investimento_edu['CO_MUNICIPIO_IBGE'] = investimento_edu['CO_MUNICIPIO_IBGE'].apply(lambda x: pd.to_numeric(x))
```

```python
investimento_edu.CO_MUNICIPIO_IBGE.str.contains('\D', regex=True).any()
```

```python
investimento_edu_df = investimento_edu[~ investimento_edu['CO_MUNICIPIO_IBGE'] \
                                       .str.contains('\D', regex=True, na=False)]
investimento_edu_df.head()
```

```python
investimento_edu_df.CO_MUNICIPIO_IBGE.str.contains('\D', regex=True).any()
```

#### Convertendo CO_MUNICIPIO_IBGE em dado numerico

```python
investimento_edu_df['CO_MUNICIPIO_IBGE'] = investimento_edu_df['CO_MUNICIPIO_IBGE'] \
                                        .apply(lambda x: pd.to_numeric(x))
investimento_edu_df.info()
```

```python
investimento = investimento_edu_df[investimento_edu_df['TP_PERIODO'] == 'ANUAL']

investimento = investimento[['CO_MUNICIPIO_IBGE', 'VL_RECEITA_REALIZADA', 
                             'VL_RECEITA_ORCADA', 'VL_DESPESA_EMPENHADA_EDUCACAO', 
                             'VL_DESPESA_LIQUIDADA_EDUCACAO','VL_DESPESA_PAGA_EDUCACAO', 
                             'VL_DESPESA_ORCADA_EDUCACAO']]
investimento.head()
```

```python
investimento.describe()
```

```python
investimento.to_csv(f'{diretorio_datasets}/investimento_municipio.csv', sep=';', encoding='utf-8', index=False)
```
